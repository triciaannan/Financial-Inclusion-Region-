```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
savings=pd.read_excel('Indicators 2000-2022.xls',header=3)
savings.head(1)
```

```python
savings=savings.set_index(['Country Name','Country Code', 'Indicator Name','Indicator Code'])
savings.head(2)
```

# Data Cleaning and Structuring

```python
savings_new= savings.stack()
savings_new.head()
```

```python
savings_new=savings_new.unstack(['Indicator Name'])
savings_new.head()
```

```python
savings_new.reset_index(inplace=True)
savings_new.head()
```

```python
savings_new.rename(columns={'level_3': 'Year'},inplace=True)
savings_new.head()
```

```python
savings_new.set_index(['Country Name','Year'],inplace= True)
savings_new.tail(5)
```

```python

regional = savings_new.loc[savings_new.index.get_level_values('Country Name').isin(['East Asia & Pacific (excluding high income)',
                                                                                        'Europe & Central Asia (excluding high income)',
                                                                                        'High income',
                                                                                        'Latin America & Caribbean (excluding high income)',
                                                                                        'Middle East & North Africa (excluding high income)',
                                                                                        'South Asia',
                                                                                        'Sub-Saharan Africa (excluding high income)'])]

regional.head()
```

```python
region=regional.reset_index()
region.head()
```

```python
East_asia=region[region['Country Name'] == 'East Asia & Pacific (excluding high income)']
Euro_asia=region[region['Country Name'] == 'Europe & Central Asia (excluding high income)']
High_income=region[region['Country Name'] == 'High income']
Latin=region[region['Country Name'] == 'Latin America & Caribbean (excluding high income)']
MENA=region[region['Country Name'] == 'Middle East & North Africa (excluding high income)']
South_Asia=region[region['Country Name'] == 'South Asia']
SSA=region[region['Country Name'] == 'Sub-Saharan Africa (excluding high income)']

print(East_asia.head(),'\n',Euro_asia.head(),'\n',High_income.head(),'\n',Latin.head(),'\n',MENA.head(),'\n',South_Asia.head(),'\n',SSA.head())
```

```python
East_asia.set_index('Year', inplace=True)
East_asia.sort_index(inplace=True)
East_asia = East_asia.loc['2000':]
East_asia.head()
```

```python
SSA.set_index('Year', inplace=True)
SSA.sort_index(inplace=True)
SSA = SSA.loc['2000':]
SSA.tail()
```

```python
High_income.set_index('Year', inplace=True)
High_income.sort_index(inplace=True)
High_income = High_income.loc['2000':]
High_income.tail()
```

```python
Latin.set_index('Year', inplace=True)
Latin.sort_index(inplace=True)
Latin = Latin.loc['2000':]
Latin.tail()
```

```python
MENA.set_index('Year', inplace=True)
MENA.sort_index(inplace=True)
MENA = MENA.loc['2000':]
MENA.tail()
```

```python
South_Asia.set_index('Year', inplace=True)
South_Asia.sort_index(inplace=True)
South_Asia = South_Asia.loc['2000':]
South_Asia.tail()
```

```python
Euro_asia.set_index('Year', inplace=True)
Euro_asia.sort_index(inplace=True)
Euro_asia = Euro_asia.loc['2000':]
Euro_asia.tail()
```

```python
fig, ax = plt.subplots(figsize=(16,4)) 

rect = fig.patch
rect.set_facecolor('#F2DFCE')

regions = [East_asia, SSA, High_income, Latin, MENA, South_Asia, Euro_asia]
colors = ['royalblue', 'salmon', 'teal', 'purple', 'gold', 'slategray', 'orange']
labels = ['EAP', 'SSA', 'HIC', 'LAC', 'MENA', 'SAS', 'ECA']

for region, color, label in zip(regions, colors, labels):
    plt.plot(region.index, region['Gross domestic savings (% of GDP)'], color=color, label=label)

plt.xlabel('', fontsize=14, color='black')
plt.ylabel('GSP', fontsize=14, color='slategray',fontweight='bold')
plt.title('Gross Domestic Savings Across Global Regions', fontsize=16, color='slategray',fontweight="bold")
plt.xticks(rotation=45)

ax.set_facecolor('#F2DFCE')
plt.legend(frameon=False,loc="lower right",ncol=7)
#plt.grid(axis='y')
plt.xlim('2000','2022')
plt.ylim(0,)
sns.despine(ax=ax)
plt.show()
```

```python
acc_own=pd.read_excel('account ownership.xlsx',header=3)
acc_own
```

```python
acc_global = acc_own['Country Name'].isin(['High income', 'Sub-Saharan Africa (excluding high income)', 'South Asia', 'Middle East & North Africa (excluding high income)', 'Latin America & Caribbean (excluding high income)', 'Europe & Central Asia (excluding high income)', 'East Asia & Pacific (excluding high income)'])
acc_globe= acc_own[acc_global]
acc_globe
```

```python
acc_globe.loc[:, 'Change'] = acc_globe['2021'] / acc_globe['2017']-1
acc_globe.head()
```

```python
acc_globe.set_index('Country Name',inplace= True)
```

```python
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(15,5)) 

y_range=np.arange(1,len(acc_globe.index)+1)
colors = np.where(acc_globe['2021'] >acc_globe['2017'],'#d9d9d9', '#d57883')


plt.hlines(y=y_range,xmin=acc_globe['2017'],xmax=acc_globe['2021'],
           colors=colors,lw=10)
plt.scatter(acc_globe['2017'],y_range,color='#0096d7', s=200,label='2017',zorder=3)
plt.scatter(acc_globe['2021'],y_range,color='#003953', s=200,label='2021',zorder=3)

for (_,row),y in zip(acc_globe.iterrows(),y_range):
    plt.annotate(f"{row['Change']:+.0%}", (max(row['2017'],row['2021']) +1,y-0.11))
    plt.legend (ncol=2,bbox_to_anchor=(1.,1.01), loc="lower right",frameon=False)

plt.yticks(y_range, acc_globe.index)
plt.title("Account ownership at a financial institution or with a mobile-money-service provider,\nsecondary education or more (% of population ages 15+) ", loc='left')
plt.xlim(50, 120)
plt.gcf().subplots_adjust(left=0.35)
plt.tight_layout()
plt.show()
```
